```python
def Fill_Age(data):
    Age=data[0]
    sex=data[1]
    if pd.isnull(Age):
        if sex is 'male':
            return 29
        else:
            return 24
    else:
        return Age
```

```python
sns.heatmap(titanic_df.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
titanic_df['Age'] = titanic_df[['Age', 'Sex']].apply(Fill_Age, axis=1)
```

```python
pd.get_dummies(titanic_df['Sex'])
```

```python
male=pd.get_dummies(titanic_df['Sex'],drop_first=True)
titanic_df.drop(['Sex'],axis=1,inplace=True)
```

```python
titanic_df
```

```python
titanic_df = pd.concat([titanic_df, male], axis=1).astype(float)
titanic_df
```

```python
x=titanic_df.drop('Survived',axis=1).values
y=titanic_df['Survived'].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0,max_iter=5000)
classifier.fit(x_train, y_train)
```

```python
y_predict_test = classifier.predict(x_test)
y_predict_test
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict_test))
```

```python
from sklearn.metrics import accuracy_score
```

```python
classifier = MultinomialNB()
classifier.fit(x_train, y_train)
```

```python
y_predict_test = classifier.predict(x_test)
cm=confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True, fmt="d")
```
